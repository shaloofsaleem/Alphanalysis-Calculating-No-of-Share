## Accept input from user for data period and investment amount And Use tradingview/Yahoo finance API 

```python
def getInput():
    amount = int(input("Enter your investment"))
    startDate = str(input("Enter Your Start Date as yyyy-mm-dd :"))
    endDate = str(input("Enter Your End Date as yyyy-mm-dd :"))
    
    return amount,startDate,endDate


def getClosp(stocks,start,end):
    
    s =yf.download(stocks,start,end)['Close']
    return s
```

## Calculate number of shares that can be bought with this amount by dividing it with daily closing price. You will get number of shares that can be bought for that stock for that particular day.

```python
number_of_shares = []
for i in range(len(close_price)):
    for j in range(len(close_price.columns)):
        amount_to_invest = investment_amount * weightage[i]
        number_share = amount_to_invest/close_price.iloc[i, j]
        number_of_shares.append(math.floor(number_share))
        
number_of_shares
```

## Change the all Code Into Standerd code With the help of Function

```python
def stock_report():
    
    data=pd.read_csv("Stocks.csv")
    stocks = list(data["Ticker"].values)
    ticker = [stock+ ".NS" for stock in stocks]
    weightage = list(data["Weightage"].values)
    portfolio = {}

    for i in range(len(data)):
        portfolio[stocks[i]]=weightage[i]
        
    def getInput():
        amount = int(input("Enter your investment"))
        startDate = str(input("Enter Your Start Date as yyyy-mm-dd :"))
        endDate = str(input("Enter Your End Date as yyyy-mm-dd :"))
    
        return amount,startDate,endDate
    
    investment_amount, start_date, end_date = getInput()

    def getClosp(stocks,start,end):
        s =yf.download(stocks,start,end)['Close']
        return s
    
    closPrice=getClosp(ticker,start=start_date ,end=end_date)
    
    close_price = closPrice.transpose()
    print(close_price)
    number_of_shares = []
    for i in range(len(close_price)):
        for j in range(len(close_price.columns)):
            amount_to_invest = investment_amount * weightage[i]
            number_share = amount_to_invest/close_price.iloc[i, j]
            number_of_shares.append(math.floor(number_share))
            
    date_strings = [date.strftime('%Y-%m-%d') for date in close_price.columns[1:]]
    columns = ["Stocks"] + date_strings
    output = {}
    idx = 0
    for i in range(len(stocks)):
        output[stocks[i]] = []
        
        for j in range(len(date_strings)):
            output[stocks[i]].append(number_of_shares[idx])
            idx+=1
    report = pd.DataFrame(output).transpose().reset_index()
    report.columns = columns
    report.to_csv('report.csv', index=False)
```
